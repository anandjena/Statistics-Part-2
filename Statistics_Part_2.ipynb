1.What is hypothesis testing in statistics?
- Hypothesis testing in statistics is a method used to make decisions or inferences about population parameters based on sample data. It involves formulating a null hypothesis (H₀) and an alternative hypothesis (H₁), then using statistical evidence to determine whether to reject the null hypothesis.

2.What is the null hypothesis, and how does it differ from the alternative hypothesis?
- The null hypothesis (H₀) is a statement that there is no effect or no difference, and it serves as the default assumption in hypothesis testing. The alternative hypothesis (H₁) is the claim that there is an effect or a difference. The goal of hypothesis testing is to determine which hypothesis is more likely given the data.

3.What is the significance level in hypothesis testing, and why is it important?
- The significance level, denoted by alpha (α), is the threshold for rejecting the null hypothesis. Commonly set at 0.05, it represents a 5% risk of concluding that an effect exists when there is none (Type I error). It is important because it defines the level of evidence needed to consider results statistically significant.

4.What does a P-value represent in hypothesis testing?
- A P-value in hypothesis testing measures the probability of obtaining results at least as extreme as those observed, assuming the null hypothesis is true. It helps assess the strength of evidence against the null hypothesis.

5.How do you interpret the P-value in hypothesis testing?
- To interpret a P-value, compare it to the significance level (α). If the P-value is less than α, the result is statistically significant, and the null hypothesis is rejected. If the P-value is greater, we fail to reject the null hypothesis.

6.What are Type 1 and Type 2 errors in hypothesis testing?
- Type I error occurs when the null hypothesis is wrongly rejected (false positive), while Type II error happens when the null hypothesis is not rejected when it is false (false negative). Managing these errors is essential for reliable statistical conclusions.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
- A one-tailed test evaluates the direction of an effect (e.g., whether a parameter is greater than a certain value), while a two-tailed test checks for any difference without specifying the direction. Two-tailed tests are more conservative and commonly used unless there is a strong rationale for a directional hypothesis.

8.What is the Z-test, and when is it used in hypothesis testing?
- The Z-test is used when the population variance is known or the sample size is large (typically n > 30). It is based on the standard normal distribution and is useful for testing hypotheses about means or proportions.

9.How do you calculate the Z-score, and what does it represent in hypothesis testing?
- The Z-score measures how many standard deviations a data point is from the mean. It is calculated by subtracting the mean from the observed value and dividing by the standard deviation. In hypothesis testing, Z-scores help determine how extreme a sample result is under the null hypothesis.

10.What is the T-distribution, and when should it be used instead of the normal distribution?
- The T-distribution is similar to the normal distribution but has heavier tails, making it suitable for small sample sizes or when the population variance is unknown. It is preferred over the normal distribution in such cases because it provides a more accurate estimate of variability.

11.What is the difference between a Z-test and a T-test?
- The difference between a Z-test and a T-test lies in the use of standard deviation: the Z-test uses known population standard deviation, while the T-test uses sample standard deviation and the T-distribution. The T-test is more appropriate for small samples.

12.What is the T-test, and how is it used in hypothesis testing?
- The T-test assesses whether the means of two groups are statistically different. It is used in various forms: one-sample, two-sample (independent), and paired T-tests, depending on the data and experimental design.

13.What is the relationship between Z-test and T-test in hypothesis testing?
- The relationship between Z-test and T-test is that both compare sample means to a population mean or to each other. The main distinction is the use of distributions: Z for known population variance and large samples; T for unknown variance and small samples.

14.What is a confidence interval, and how is it used to interpret statistical results?
- A confidence interval is a range of values derived from sample data that is likely to contain the true population parameter. It helps interpret statistical results by providing a measure of precision and uncertainty.

15.What is the margin of error, and how does it affect the confidence interval?
- The margin of error is the amount added and subtracted from the sample estimate to form a confidence interval. A smaller margin of error indicates more precise estimates. It is influenced by sample size, variability, and confidence level.

16.How is Bayes' Theorem used in statistics, and what is its significance?
- Bayes' Theorem is used to update the probability of a hypothesis as more evidence becomes available. In statistics, it underpins Bayesian inference, allowing for the incorporation of prior beliefs in probability estimates.

17.What is the Chi-square distribution, and when is it used?
- The Chi-square distribution is used primarily in tests of independence and goodness of fit. It helps assess whether observed frequencies differ significantly from expected frequencies.

18.What is the Chi-square goodness of fit test, and how is it applied?
- The Chi-square goodness of fit test evaluates how well observed categorical data match expected distributions. It is applied when testing hypotheses about the distribution of a single categorical variable.

19.What is the F-distribution, and when is it used in hypothesis testing?
- The F-distribution arises in the context of comparing variances and is used in ANOVA and regression analysis. It helps test whether group variances are equal or if models explain significant variation.

20.What is an ANOVA test, and what are its assumptions?
- An ANOVA test (Analysis of Variance) determines whether there are significant differences among the means of three or more groups. It assumes normal distribution, equal variances, and independent samples.

21.What are the different types of ANOVA tests?
- The different types of ANOVA include one-way ANOVA (testing one factor), two-way ANOVA (testing two factors), and repeated measures ANOVA (for correlated samples over time or conditions).

22.What is the F-test, and how does it relate to hypothesis testing?
- The F-test compares the variances of two or more groups and is closely related to ANOVA. It helps determine if observed group differences are statistically significant based on the ratio of variances.

```python
# 1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
from scipy.stats import norm

# Inputs
sample = [54, 49, 51, 52, 50, 53, 48, 52, 49, 51]  # Sample data
population_mean = 50  # Known population mean
population_std = 2    # Known population standard deviation
alpha = 0.05          # Significance level

# Calculate sample statistics
sample_mean = np.mean(sample)
sample_size = len(sample)
standard_error = population_std / np.sqrt(sample_size)

# Calculate Z-score
z_score = (sample_mean - population_mean) / standard_error

# Calculate P-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Output results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-score = {z_score:.3f}")
print(f"P-value = {p_value:.4f}")

# Interpretation
if p_value < alpha:
    print("Result: Reject the null hypothesis. The sample mean is significantly different from the population mean.")
else:
    print("Result: Fail to reject the null hypothesis. The sample mean is not significantly different from the population mean.")
```

```python
# 4.Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
population_mean = 100
population_std = 15
sample_size = 30
alpha = 0.05  # significance level

# Simulate sample data
np.random.seed(0)
sample = np.random.normal(loc=105, scale=15, size=sample_size)
sample_mean = np.mean(sample)
standard_error = population_std / np.sqrt(sample_size)

# Z-test
z_score = (sample_mean - population_mean) / standard_error
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Critical z-values for two-tailed test
z_critical = norm.ppf(1 - alpha/2)

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.3f}")
print(f"P-value: {p_value:.4f}")

if abs(z_score) > z_critical:
    print("Result: Reject the null hypothesis.")
else:
    print("Result: Fail to reject the null hypothesis.")

# Visualization
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution')

# Shade the rejection regions
plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.4, label='Rejection Region (Left)')
plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.4, label='Rejection Region (Right)')

# Plot the observed Z-score
plt.axvline(z_score, color='blue', linestyle='--', linewidth=2, label=f'Z-score = {z_score:.2f}')

# Labels and legend
plt.title('Two-Tailed Z-Test with Rejection Regions')
plt.xlabel('Z-value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0, mu1, sigma, n, alpha=0.05):
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value and cutoff point under H0
    z_critical = norm.ppf(1 - alpha)
    critical_value = mu0 + z_critical * se

    # Type II error: P(fail to reject H0 | H1 is true)
    beta = norm.cdf((critical_value - mu1) / se)

    # X range for plotting
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    y_null = norm.pdf(x, mu0, se)
    y_alt = norm.pdf(x, mu1, se)

    # Plot distributions
    plt.figure(figsize=(10, 6))
    plt.plot(x, y_null, label='H₀ Distribution', color='blue')
    plt.plot(x, y_alt, label='H₁ Distribution', color='green')

    # Shade Type I error region (right tail under H₀)
    plt.fill_between(x, y_null, where=(x > critical_value), color='red', alpha=0.4, label='Type I Error (α)')

    # Shade Type II error region (left of critical under H₁)
    plt.fill_between(x, y_alt, where=(x <= critical_value), color='orange', alpha=0.4, label='Type II Error (β)')

    # Annotations and aesthetics
    plt.axvline(critical_value, color='black', linestyle='--', label=f'Critical Value = {critical_value:.2f}')
    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print error probabilities
    print(f"Type I Error (α): {alpha:.2f}")
    print(f"Type II Error (β): {beta:.2f}")
    print(f"Power of the Test (1 - β): {1 - beta:.2f}")

# Example usage
visualize_type1_type2_errors(mu0=100, mu1=105, sigma=15, n=30, alpha=0.05)
```

```python
# 6.Write a Python program to perform an independent T-test and interpret the results
import numpy as np
from scipy.stats import ttest_ind

# Set random seed for reproducibility
np.random.seed(1)

# Simulate two independent samples
group_a = np.random.normal(loc=100, scale=10, size=30)  # mean = 100
group_b = np.random.normal(loc=105, scale=10, size=30)  # mean = 105

# Perform independent two-sample t-test
t_stat, p_value = ttest_ind(group_a, group_b, equal_var=True)  # assume equal variance

# Output results
print(f"Group A mean: {np.mean(group_a):.2f}")
print(f"Group B mean: {np.mean(group_b):.2f}")
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis. The group means are significantly different.")
else:
    print("Result: Fail to reject the null hypothesis. No significant difference between group means.")
```

```python
# 9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance
import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
        sample (array-like): The sample data.
        confidence (float): Confidence level (default is 0.95 for 95%).

    Returns:
        tuple: (lower_bound, upper_bound) of the confidence interval
    """
    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)
    t_critical = t.ppf((1 + confidence) / 2, df=n - 1)

    margin_of_error = t_critical * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound


# Simulated sample data
np.random.seed(1)
data = np.random.normal(loc=50, scale=10, size=25)

# Compute 95% confidence interval
ci_low, ci_high = confidence_interval(data)
print(f"95% Confidence Interval: ({ci_low:.2f}, {ci_high:.2f})")
```

```python
# 10.Write a Python program to calculate the margin of error for a given confidence level using sample data
import numpy as np
from scipy.stats import t

def margin_of_error(sample, confidence=0.95):
    """
    Calculate the margin of error for a sample mean.

    Parameters:
        sample (array-like): The sample data.
        confidence (float): The desired confidence level (default 0.95).

    Returns:
        float: Margin of error
    """
    n = len(sample)
    sample_std = np.std(sample, ddof=1)  # sample standard deviation
    std_error = sample_std / np.sqrt(n)
    t_critical = t.ppf((1 + confidence) / 2, df=n - 1)
    moe = t_critical * std_error
    return moe

# Simulate sample data
np.random.seed(42)
data = np.random.normal(loc=75, scale=8, size=30)

# Calculate margin of error at 95% confidence
moe_95 = margin_of_error(data, confidence=0.95)
print(f"Margin of Error (95% confidence): ±{moe_95:.2f}")
```

```python
# 11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
def bayes_theorem(prior_H, likelihood_E_given_H, likelihood_E_given_not_H):
    """
    Implements Bayes' Theorem to compute the posterior probability.

    Parameters:
        prior_H: P(H) - prior probability of the hypothesis
        likelihood_E_given_H: P(E|H) - probability of evidence if hypothesis is true
        likelihood_E_given_not_H: P(E|¬H) - probability of evidence if hypothesis is false

    Returns:
        P(H|E): posterior probability
    """
    # Prior of not H
    prior_not_H = 1 - prior_H

    # Total probability of evidence (P(E))
    evidence_prob = (likelihood_E_given_H * prior_H) + (likelihood_E_given_not_H * prior_not_H)

    # Bayes' theorem
    posterior = (likelihood_E_given_H * prior_H) / evidence_prob
    return posterior


# Scenario: Disease testing
prior = 0.01               # 1% have the disease
sensitivity = 0.99         # P(Test+ | Disease)
false_positive_rate = 0.05 # P(Test+ | No Disease)

# Calculate posterior
posterior_prob = bayes_theorem(prior, sensitivity, false_positive_rate)
print(f"Probability of having the disease given a positive test: {posterior_prob:.4f}")
```

```python
# 12.Perform a Chi-square test for independence between two categorical variables in Python.
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Create a contingency table (observed frequencies)
# Rows: Gender, Columns: Product Preference
data = pd.DataFrame({
    'Product A': [30, 20],   # [Males, Females]
    'Product B': [10, 40]
}, index=['Male', 'Female'])

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(data)

# Print the results
print("Contingency Table (Observed):")
print(data)
print("\nExpected Frequencies (if independent):")
print(pd.DataFrame(expected, index=data.index, columns=data.columns))

print(f"\nChi-square statistic: {chi2:.3f}")
print(f"Degrees of freedom: {dof}")
print(f"P-value: {p:.4f}")

# Interpretation
alpha = 0.05
if p < alpha:
    print("Result: Reject the null hypothesis. There is a significant association between the variables.")
else:
    print("Result: Fail to reject the null hypothesis. No significant association found.")
```

```python
# 13.Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test.

    Parameters:
        observed (2D list or DataFrame): Contingency table of observed frequencies

    Returns:
        expected (DataFrame): Expected frequencies under the assumption of independence
    """
    # Convert to DataFrame if needed
    if not isinstance(observed, pd.DataFrame):
        observed = pd.DataFrame(observed)

    row_totals = observed.sum(axis=1)
    col_totals = observed.sum(axis=0)
    grand_total = observed.values.sum()

    # Outer product of row and column totals divided by grand total
    expected = np.outer(row_totals, col_totals) / grand_total

    # Format expected frequencies as a DataFrame
    expected_df = pd.DataFrame(expected, index=observed.index, columns=observed.columns)
    return expected_df

# Example observed data: Gender vs. Product Preference
observed_data = pd.DataFrame({
    'Product A': [30, 20],   # [Males, Females]
    'Product B': [10, 40]
}, index=['Male', 'Female'])

# Calculate expected frequencies
expected_freq = calculate_expected_frequencies(observed_data)

# Display results
print("Observed Frequencies:")
print(observed_data)
print("\nExpected Frequencies (if variables are independent):")
print(expected_freq.round(2))
```

```python
# 14.Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution
import numpy as np
from scipy.stats import chisquare

# Observed frequencies (e.g., counts from rolling a die 60 times)
observed = np.array([10, 9, 12, 11, 8, 10])

# Expected frequencies for a fair die (uniform distribution)
expected = np.full_like(observed, fill_value=np.sum(observed) / len(observed))

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Output results
print("Observed Frequencies:", observed)
print("Expected Frequencies:", expected)
print(f"\nChi-square Statistic: {chi2_stat:.3f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis. Observed data does not fit the expected distribution.")
else:
    print("Result: Fail to reject the null hypothesis. Data fits the expected distribution.")
```

```python
# 16.Implement an F-test using Python to compare the variances of two random samples.
import numpy as np
from scipy.stats import f

def f_test(sample1, sample2, alpha=0.05):
    """
    Performs an F-test to compare the variances of two samples.

    Parameters:
        sample1, sample2: array-like numeric samples
        alpha: significance level (default 0.05)

    Returns:
        F statistic, p-value, test conclusion
    """
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)
    df1 = len(sample1) - 1
    df2 = len(sample2) - 1

    # Ensure F >= 1 by placing the larger variance in the numerator
    if var1 > var2:
        F = var1 / var2
        dfn, dfd = df1, df2
    else:
        F = var2 / var1
        dfn, dfd = df2, df1

    # Two-tailed p-value
    p_value = 2 * min(f.cdf(F, dfn, dfd), 1 - f.cdf(F, dfn, dfd))

    conclusion = "Reject H0: Variances are significantly different" if p_value < alpha else "Fail to reject H0: No significant difference in variances"

    return F, p_value, conclusion

# Simulate two random samples
np.random.seed(42)
sample_a = np.random.normal(loc=10, scale=2, size=30)
sample_b = np.random.normal(loc=10, scale=3, size=30)

# Perform F-test
F_stat, p_val, result = f_test(sample_a, sample_b)

print(f"F-statistic: {F_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(f"Result: {result}")
```

```python
# 19.Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

import numpy as np
import pandas as pd
from scipy.stats import shapiro, levene

def check_anova_assumptions(groups):
    """
    Checks normality and homogeneity of variances for ANOVA.

    Parameters:
        groups (list of arrays or Series): List of numerical samples for each group.

    Returns:
        Dictionary summarizing test results.
    """
    results = {}

    # Normality check (Shapiro-Wilk test)
    print("Normality Test (Shapiro-Wilk):")
    normality = []
    for i, group in enumerate(groups):
        stat, p = shapiro(group)
        normality.append(p > 0.05)
        print(f"  Group {i+1}: W = {stat:.4f}, p = {p:.4f} → {'Normal' if p > 0.05 else 'Not normal'}")

    # Homogeneity of variances (Levene’s Test)
    print("\nEqual Variance Test (Levene's):")
    stat, p = levene(*groups)
    equal_var = p > 0.05
    print(f"  Levene's test: W = {stat:.4f}, p = {p:.4f} → {'Equal variances' if equal_var else 'Unequal variances'}")

    results['all_normal'] = all(normality)
    results['equal_variances'] = equal_var
    return results

# Simulated group data
np.random.seed(1)
group1 = np.random.normal(50, 5, 30)
group2 = np.random.normal(52, 5, 30)
group3 = np.random.normal(48, 5, 30)

# Check ANOVA assumptions
assumption_results = check_anova_assumptions([group1, group2, group3])

# Decision
if assumption_results['all_normal'] and assumption_results['equal_variances']:
    print("\n✅ Assumptions met. Safe to proceed with ANOVA.")
else:
    print("\n⚠️ One or more assumptions violated. Consider using a non-parametric test (e.g., Kruskal-Wallis).")
```

```python
# 20.Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
import matplotlib.pyplot as plt

# Simulate data
np.random.seed(42)
n = 20
data = pd.DataFrame({
    'Factor_A': np.repeat(['Low', 'High'], n * 2),
    'Factor_B': np.tile(np.repeat(['Type1', 'Type2'], n), 2),
    'Score': np.concatenate([
        np.random.normal(50, 5, n),   # A=Low, B=Type1
        np.random.normal(55, 5, n),   # A=Low, B=Type2
        np.random.normal(60, 5, n),   # A=High, B=Type1
        np.random.normal(65, 5, n)    # A=High, B=Type2
    ])
})

# Perform two-way ANOVA
model = ols('Score ~ C(Factor_A) + C(Factor_B) + C(Factor_A):C(Factor_B)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Display results
print("Two-Way ANOVA Results")
print(anova_table)

# Visualization of interaction
plt.figure(figsize=(8, 6))
sns.pointplot(data=data, x='Factor_A', y='Score', hue='Factor_B', capsize=.1, palette='Set2')
plt.title('Interaction Plot: Factor A vs. Factor B')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 21. Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Define degrees of freedom
df1 = 5   # numerator degrees of freedom
df2 = 20  # denominator degrees of freedom

# Generate x values
x = np.linspace(0, 5, 500)

# Compute F-distribution PDF
y = f.pdf(x, df1, df2)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'F-distribution (df1={df1}, df2={df2})', color='darkblue')
plt.fill_between(x, y, alpha=0.2, color='skyblue')

# Mark critical value at 0.95 (5% significance level)
alpha = 0.05
f_crit = f.ppf(1 - alpha, df1, df2)
plt.axvline(f_crit, color='red', linestyle='--', label=f'Critical Value = {f_crit:.2f}')

# Annotate
plt.title("F-Distribution with df1 = 5 and df2 = 20")
plt.xlabel("F-value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 23.Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means
import numpy as np
from scipy.stats import ttest_1samp
import matplotlib.pyplot as plt

# Simulate data
np.random.seed(42)
sample_size = 50
true_mean = 100
data = np.random.normal(loc=true_mean, scale=10, size=sample_size)

# Hypothesized population mean (μ₀)
mu_0 = 105

# Perform one-sample t-test
t_stat, p_value = ttest_1samp(data, mu_0)

# Print test result
print("One-Sample T-Test")
print("------------------")
print(f"Sample Mean: {np.mean(data):.2f}")
print(f"Hypothesized Mean: {mu_0}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject H₀ — sample mean is significantly different from the hypothesized mean.")
else:
    print("Conclusion: Fail to reject H₀ — no significant difference between sample mean and hypothesized mean.")

# Optional: Plot histogram of the sample data
plt.figure(figsize=(8, 5))
plt.hist(data, bins=10, color='skyblue', edgecolor='black')
plt.axvline(mu_0, color='red', linestyle='--', label=f'Hypothesized μ₀ = {mu_0}')
plt.axvline(np.mean(data), color='blue', linestyle='-', label=f'Sample Mean = {np.mean(data):.2f}')
plt.title("Sample Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 24.Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results
import numpy as np
from scipy.stats import chi2

# Simulate data from a normal population
np.random.seed(42)
sample = np.random.normal(loc=100, scale=10, size=30)

# Known population variance to test against
sigma_squared_0 = 100  # i.e., σ₀² = 10²
n = len(sample)
sample_variance = np.var(sample, ddof=1)

# Compute chi-square test statistic
chi2_stat = (n - 1) * sample_variance / sigma_squared_0

# Degrees of freedom
df = n - 1

# Critical values for a two-tailed test at alpha = 0.05
alpha = 0.05
chi2_critical_low = chi2.ppf(alpha / 2, df)
chi2_critical_high = chi2.ppf(1 - alpha / 2, df)

# Print results
print("Chi-square Test for Population Variance")
print("----------------------------------------")
print(f"Sample Variance: {sample_variance:.2f}")
print(f"Hypothesized Variance (σ₀²): {sigma_squared_0}")
print(f"Chi-square statistic: {chi2_stat:.2f}")
print(f"Critical values (α = 0.05): [{chi2_critical_low:.2f}, {chi2_critical_high:.2f}]")

if chi2_stat < chi2_critical_low or chi2_stat > chi2_critical_high:
    print("Conclusion: Reject H₀ — population variance is significantly different from σ₀².")
else:
    print("Conclusion: Fail to reject H₀ — no significant difference in population variance.")
```

```python
# 25.Write a Python script to perform a Z-test for comparing proportions between two datasets or groups
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Sample data: [successes, sample size]
successes = np.array([60, 45])     # e.g., number of people who responded positively
samples = np.array([100, 90])      # e.g., total people surveyed in each group

# Perform two-proportion Z-test
z_stat, p_value = proportions_ztest(successes, samples)

# Print results
print("Two-Proportion Z-Test")
print("----------------------")
print(f"Group 1: {successes[0]}/{samples[0]} = {successes[0]/samples[0]:.2f}")
print(f"Group 2: {successes[1]}/{samples[1]} = {successes[1]/samples[1]:.2f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject H₀ — proportions are significantly different.")
else:
    print("Conclusion: Fail to reject H₀ — no significant difference in proportions.")
```

```python
# 26.Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results
import numpy as np
from scipy.stats import f
import matplotlib.pyplot as plt

# Simulate two datasets
np.random.seed(42)
data1 = np.random.normal(loc=50, scale=5, size=30)
data2 = np.random.normal(loc=50, scale=8, size=30)

# Calculate sample variances
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

# Ensure larger variance is numerator (F >= 1)
if var1 > var2:
    f_stat = var1 / var2
    dfn, dfd = len(data1) - 1, len(data2) - 1
else:
    f_stat = var2 / var1
    dfn, dfd = len(data2) - 1, len(data1) - 1

# Compute critical values for two-tailed test
alpha = 0.05
f_crit_low = f.ppf(alpha / 2, dfn, dfd)
f_crit_high = f.ppf(1 - alpha / 2, dfn, dfd)

# Print F-test result
print("F-Test for Equality of Variances")
print("---------------------------------")
print(f"Variance 1: {var1:.2f}")
print(f"Variance 2: {var2:.2f}")
print(f"F-statistic: {f_stat:.4f}")
print(f"Critical values at α = 0.05: [{f_crit_low:.4f}, {f_crit_high:.4f}]")

if f_stat < f_crit_low or f_stat > f_crit_high:
    print("Conclusion: Reject H₀ — variances are significantly different.")
else:
    print("Conclusion: Fail to reject H₀ — no significant difference in variances.")

# Visualization
x = np.linspace(0.1, 5, 500)
y = f.pdf(x, dfn, dfd)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='F-distribution', color='darkblue')
plt.axvline(f_stat, color='red', linestyle='--', label=f'F-stat = {f_stat:.2f}')
plt.axvline(f_crit_low, color='gray', linestyle='--', label='Critical values')
plt.axvline(f_crit_high, color='gray', linestyle='--')
plt.fill_between(x, y, where=(x < f_crit_low) | (x > f_crit_high), color='orange', alpha=0.3, label='Rejection Region')
plt.title("F-distribution and Test Statistic")
plt.xlabel("F-value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results
import numpy as np
from scipy.stats import chisquare
import matplotlib.pyplot as plt

# Simulate observed data: counts from a categorical outcome
np.random.seed(42)
observed = np.random.multinomial(n=100, pvals=[0.2, 0.3, 0.3, 0.2])
categories = ['A', 'B', 'C', 'D']

# Define expected frequencies (e.g., uniform or a specific known distribution)
expected_probs = [0.25, 0.25, 0.25, 0.25]
expected = np.array(expected_probs) * sum(observed)

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Print results
print("Chi-Square Goodness-of-Fit Test")
print("--------------------------------")
print(f"Observed counts: {observed}")
print(f"Expected counts: {expected.astype(int)}")
print(f"Chi-square statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject H₀ — observed data does not fit the expected distribution.")
else:
    print("Conclusion: Fail to reject H₀ — observed data fits the expected distribution.")

# Plot observed vs expected
x = np.arange(len(categories))
bar_width = 0.35

plt.figure(figsize=(8, 6))
plt.bar(x - bar_width/2, observed, width=bar_width, label='Observed', color='skyblue')
plt.bar(x + bar_width/2, expected, width=bar_width, label='Expected', color='salmon')
plt.xticks(x, categories)
plt.title('Observed vs. Expected Frequencies')
plt.ylabel('Count')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
